# 1. Fetch S&P 500 data

Requirements

```python
# Imports
import requests
import pandas as pd
import os
```

```python
# Define and prepare destination directory
DATA_DIR = '../data/'
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
# Fetch S&P 500 Companies and preview data
URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(URL, header=0)
table = tables[0]
table.head()
```

```python
# Reduce, save and preview data
companies_data = table[['Symbol', 'Security']]
companies_data.to_csv(DATA_DIR + 'SP500_symbols_names.csv')
companies_data.head()
```
